```python
#Here there is two null values ('not available')& ('unknown') are present but it is not treated as null in data
# I want to treat those two values as null

df=pd.read_csv('orders.csv',na_values=['Not Available','unknown']) 
df
```

```python
#Now we can see increment of null values in ship mode column

df.isnull().sum()
```

```python
#renaming column making them lowercase and replace space with underscore

df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
#Deriving new columns discount, sale price, profit

df['discount']=df['list_price']*df['discount_percent']/100
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
df.dtypes
```

```python
#converting data types of date 

df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
#dropping columns

df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```
